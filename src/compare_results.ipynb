```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

RESULTS_PATH = Path(".", "..", "data", "results")
```

```python
df_threshold = pd.read_csv(RESULTS_PATH / "threshold_rb3.2.csv")
df_threshold
```

```python
df_bayesian = pd.read_csv(RESULTS_PATH / "bayesian_rb3.2.csv")
df_bayesian.rename(columns={"SummaryErrorMetric": "error"}, inplace=True)
df_bayesian
```

```python
df_cnn = pd.read_csv(RESULTS_PATH / "cnn_results_rb3.2_10_epochs_all_traps.csv")
df_cnn.rename(columns={"SummaryErrorMetric": "error"}, inplace=True)
df_cnn
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 5), dpi=150)

data_mean_threshold = df_threshold.groupby(by="texp").error.mean()
ax.plot(
    data_mean_threshold.index,
    1 - data_mean_threshold.values / 2,
    "-o",
    label="Threshold detector",
    color="C0",
)

data_mean_bayes = df_bayesian.groupby(by="texp").error.mean()
ax.plot(
    data_mean_bayes.index,
    1 - data_mean_bayes.values / 2,
    "-o",
    label="Bayesian detector",
    color="C1",
)


data_mean_cnn = df_cnn.groupby(by="texp").error.mean()
ax.plot(
    data_mean_cnn.index,
    1 - data_mean_cnn.values / 2,
    "-o",
    label="CNN detector",
    color="C2",
)

ax.set_xlabel(r"Second image exposure time $\tau_\exp$, ms", fontsize=14)
ax.set_ylabel(
    '"Imaging fidelity" metric\n'
    + r"$\mathcal{F} = 1 - 0.5 \left[ P(\mathbf{0}|\mathbf{1}) + P(\mathbf{1}|\mathbf{0}) \right]$",
    fontsize=14
)
plt.ylim((0.989, 1.001))
ax.hlines(0.999, 0, 10, label="$\mathcal{F} = 99.9\%$", linestyles="dashed", color="red")
ax.legend(loc="lower left", fontsize=14)
ax.grid()
ax.set_xlim((0, 10))
plt.gca().tick_params(axis="both", which="major", labelsize=14)
plt.tight_layout()
plt.savefig(RESULTS_PATH / "comparison_en.png", dpi=300, bbox_inches="tight")
plt.savefig(RESULTS_PATH / "comparison_en.pdf", dpi=300, bbox_inches="tight")
plt.show()
```
